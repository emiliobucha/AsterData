```python
import pandas as pd
import numpy as np
import math
```

```python
df_asteroides_crateres = pd.read_csv('C:\Users\Emilio\Desktop\AsterData\AsteroidesCometasLimpio.csv')
df_probabilidades = pd.read_csv('C:\Users\Emilio\Desktop\AsterData\Probabilidades.csv')
```

```python
df_asteroides_crateres['Nombre']=df_asteroides_crateres['full_name'].str.strip()
```

```python
df_probabilidades['Nombre']=df_probabilidades['Object Designation  '].str.strip()
```

```python
df_merge=pd.merge(df_asteroides_crateres, df_probabilidades,left_on='Nombre', right_on='Nombre')
```

```python
df_merge['probabilidad_cerca']= df_merge['Impact Probability (cumulative)'] * 1/510072000
cols = list(['id'] + ['Nombre'] + ['pha']+['diameter']+['Estimated Diameter (km)']+['Impact Probability (cumulative)']+['probabilidad_cerca'])
df_final=df_merge[cols].sort_values(by=['probabilidad_cerca'],ascending=[False])
df_final
```

```python
data_diameter = df_final[(np.isfinite(df_final.diameter))]

velocidad = 25000;
densidad_sedimentaria = 2500

data_porous = data_diameter.sample(frac=0.75)
data_rest =data_diameter.loc[~data_diameter.index.isin(data_porous.index)]
data_dense = data_rest.sample(frac=0.48)
data_metalic = data_rest.loc[~data_rest.index.isin(data_dense.index)]
data_porous['density']=1500
data_dense['density']=3000
data_metalic['density']=8000
data_diameter_total = data_porous.append([data_dense,data_metalic])


data_diameter_total['Dtc']=1.161*np.power((data_diameter_total['density']/densidad_sedimentaria),(1./3))*np.power((data_diameter_total['diameter']*1000),0.78)*np.power(velocidad,0.44)*np.power(9.807,-0.22)*np.power(np.sin(np.pi/4),1./3)
data_diameter_total['Dfr']=1.25 * data_diameter_total['Dtc']
data_diameter_total.loc[data_diameter_total['Dtc']>2560,'Dfr']=1.17*(np.power(data_diameter_total[data_diameter_total['Dtc']>2560]['Dtc'] /1000,1.13)/np.power(3.2,0.13))
data_diameter_total['Vbr']=0.032*np.power(data_diameter_total['Dfr'],3)
data_diameter_total['dtc1']=data_diameter_total['Dtc']/(np.sqrt(2)*2)
data_diameter_total['hfr']=0.07*(np.power(data_diameter_total['Dtc'],4)/np.power(data_diameter_total['Dfr'],3))
data_diameter_total['tbr']=2.8*data_diameter_total['Vbr']*((data_diameter_total['dtc1']+data_diameter_total['hfr'])/(data_diameter_total['dtc1']*np.power(data_diameter_total['Dfr'],2)))
data_diameter_total['dfr1']=(data_diameter_total['dtc1']+data_diameter_total['hfr']-data_diameter_total['tbr'])/1000
data_diameter_total.loc[data_diameter_total['Dtc']>2560,'dfr1']=0.4*(np.power(data_diameter_total[data_diameter_total['Dtc']>2560]['Dfr'],0.3))
data_diameter_total.loc[data_diameter_total['Dtc']<2560,'Dfr']=data_diameter_total['Dfr']/1000
data_diameter_total

```

```python
data_estimated = df_final[(np.isnan(df_final.diameter))]

velocidad = 25000;
densidad_sedimentaria = 2500
angulo = np.pi/4 #Radianes

data_porous1 = data_estimated.sample(frac=0.75)
data_rest1 =data_estimated.loc[~data_estimated.index.isin(data_porous1.index)]
data_dense1 = data_rest1.sample(frac=0.48)
data_metalic1 = data_rest1.loc[~data_rest1.index.isin(data_dense1.index)]
data_porous1['density']=1500
data_dense1['density']=3000
data_metalic1['density']=8000
data_estimated_total = data_porous1.append([data_dense1,data_metalic1])


data_estimated_total['Dtc']=1.161*np.power((data_estimated_total['density']/densidad_sedimentaria),(1./3))*np.power((data_estimated_total['Estimated Diameter (km)']*1000),0.78)*np.power(velocidad,0.44)*np.power(9.807,-0.22)*np.power(np.sin(angulo),1./3)
data_estimated_total['Dfr']=1.25 * data_estimated_total['Dtc']
data_estimated_total.loc[data_estimated_total['Dtc']>2560,'Dfr']=1.17*(np.power(data_estimated_total[data_estimated_total['Dtc']>2560]['Dtc'] /1000,1.13)/np.power(3.2,0.13))
data_estimated_total['Vbr']=0.032*np.power(data_estimated_total['Dfr'],3)
data_estimated_total['dtc1']=data_estimated_total['Dtc']/(np.sqrt(2)*2)
data_estimated_total['hfr']=0.07*(np.power(data_estimated_total['Dtc'],4)/np.power(data_estimated_total['Dfr'],3))
data_estimated_total['tbr']=2.8*data_estimated_total['Vbr']*((data_estimated_total['dtc1']+data_estimated_total['hfr'])/(data_estimated_total['dtc1']*np.power(data_estimated_total['Dfr'],2)))
data_estimated_total['dfr1']=(data_estimated_total['dtc1']+data_estimated_total['hfr']-data_estimated_total['tbr'])/1000
data_estimated_total.loc[data_estimated_total['Dtc']>2560,'dfr1']=0.4*(np.power(data_estimated_total[data_estimated_total['Dtc']>2560]['Dfr'],0.3))

data_estimated_total.loc[data_estimated_total['Dtc']<2560,'Dfr']=data_estimated_total['Dfr']/1000


df_final = data_estimated_total.append(data_diameter_total)

df_final=df_final.sort_values(by=['probabilidad_cerca'],ascending=[False])
df_final
```
